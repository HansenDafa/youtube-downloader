```python
import tkinter as tk
from tkinter import messagebox, filedialog
from pytube import YouTube
import os

def download_video():
    url = url_entry.get()
    if not url:
        messagebox.showwarning("Input Error", "Please enter a YouTube URL")
        return

    download_path = filedialog.askdirectory()
    if not download_path:
        messagebox.showwarning("Input Error", "Please select a download folder")
        return

    try:
        yt = YouTube(url)
        
        if format_var.get() == "mp4":
            if quality_var.get() == "High":
                ys = yt.streams.get_highest_resolution()
            else:
                ys = yt.streams.filter(progressive=True, file_extension='mp4').order_by('resolution').asc().first()
        elif format_var.get() == "mp3":
            ys = yt.streams.filter(only_audio=True).first()

        ys.download(output_path=download_path)
        status_label.config(text=f"Download completed! Saved to {download_path}")
    except Exception as e:
        status_label.config(text="Error: " + str(e))

# Membuat jendela utama
root = tk.Tk()
root.title("YouTube Video Downloader")

# Menambahkan entry field untuk URL
url_label = tk.Label(root, text="YouTube URL:")
url_label.pack()
url_entry = tk.Entry(root, width=50)
url_entry.pack()

# Menambahkan option menu untuk memilih format
format_label = tk.Label(root, text="Select Format:")
format_label.pack()
format_var = tk.StringVar(value="mp4")
format_menu = tk.OptionMenu(root, format_var, "mp4", "mp3")
format_menu.pack()

# Menambahkan option menu untuk memilih kualitas (hanya untuk mp4)
quality_label = tk.Label(root, text="Select Quality:")
quality_label.pack()
quality_var = tk.StringVar(value="High")
quality_menu = tk.OptionMenu(root, quality_var, "High", "Low")
quality_menu.pack()

# Menambahkan tombol download
download_button = tk.Button(root, text="Download", command=download_video)
download_button.pack()

# Menambahkan label status
status_label = tk.Label(root, text="")
status_label.pack()

# Menjalankan aplikasi
root.mainloop()
```
